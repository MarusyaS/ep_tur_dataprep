## Таблица изображений
Идентификаторы изображений формируются по схеме "идентификатор модели" + буква, обозначающая тип изображения. На момент публикации подготовлены изображения трёх типов: X,Y,Z. На основании таблицы моделей формируется таблица изображений с тремя типами.
1. Создаём датафрейм
2. Добавляем в него списки идентификаторов моделей и надписей из таблицы моделей
3. Повторяем идентификаторы моделей 3 раза и добавляем к каждому одну из буквы -- типов изображения
4. Добавляем словесное описание типа в отдельной колонке

```python
# create an empty dataframe with the desired columns
image = pd.DataFrame(columns=['ID', 'Inscription_id', 'Type', 'AdditionalDescription'])

# fill the 'ID' column with values from the 'ID' column in the 'reduced_model' dataframe
image['ID'] = reduced_models['ID'].tolist()
image['Inscription_id'] = reduced_models['Inscription_id'].tolist()
# create a new column 'Variant' with three types of images 'X', 'Y', 'Z'
variants = ['X', 'Y', 'Z']
# repeat each value in the 'ID' column three times with a different variant added at the end
image = image.loc[image.index.repeat(3)]
image['Variant'] = [f'{v}' for v in variants] * len(reduced_models)

# concatenate the 'ID' column and the 'Variant' column to create a new 'ID' column with the desired format
image['ID'] = image['ID'] + image['Variant']

# fill the 'Type' column based on the variant in the 'ID' column
image.loc[image['Variant'] == 'X', 'Type'] = 'рендер общего вида в изометрии, текстурированный'
image.loc[image['Variant'] == 'Y', 'Type'] = 'рендер общего вида в изометрии, без текстуры «синий»'
image.loc[image['Variant'] == 'Z', 'Type'] = 'рендер общего вида в изометрии без текстуры «серый»'

image = image.drop(columns='Variant')
final_image = image.loc[:, ["ID","Type",'AdditionalDescription','Inscription_id']]
# image.set_index('ID')
final_image.to_csv('Images_final.csv', index=False, encoding= 'utf-8')
```
